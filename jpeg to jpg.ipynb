```python
import os

def rename_images(root_directory):
    for dirpath, dirnames, filenames in os.walk(root_directory):
        for filename in filenames:
            if filename.endswith(".jpg"):
                # Construct the full path to the image file
                image_path = os.path.join(dirpath, filename)
                
                # Construct the new filename by replacing ".jpg" with ".jpeg"
                new_filename = os.path.splitext(filename)[0] + ".jpeg"
                
                # Construct the full path to the new image file
                new_image_path = os.path.join(dirpath, new_filename)
                
                # Rename the file
                os.rename(image_path, new_image_path)
                
                print(f"Renamed: {image_path} -> {new_image_path}")

# Specify the root directory containing the folders with images
root_directory = "dataset"

# Call the function to rename images in the specified directory structure
rename_images(root_directory)
```
